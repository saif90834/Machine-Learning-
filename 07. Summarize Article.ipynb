## Pre Process Downloaded Text
Read the Article from a file/web
List of words that are present in the list  
From above list remove stopwords  
Filter-out stopwords from our article  


```python
#Read the article and store it in a variable
article = ""
with open('Data/NLP/news.txt', 'r') as file:
    data = file.read()
article = data
print(article)
```

```python
#Preprocess Downloaded Text
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation

text = article
sent_token = sent_tokenize(text)

#Convert article into list of words
word_token = word_tokenize(text.lower())

#Remove stopwords and punctuation
stop_custom=set(stopwords.words('english')+list(punctuation)+["''"])

#Take all words after removing stopwords from it
word_list=[word for word in word_token if word not in stop_custom]
word_list
```

```python
#Extracting the most common words
from nltk.probability import FreqDist
#Construct a frequency distribution of words
freq = FreqDist(word_list)
print(freq.most_common)

#Print the top 10 common words
from heapq import nlargest

print (nlargest(20, freq, key=freq.get))
```

```python

#Compute the score for each sentence (Keys -> Sentences, Values -> Significance Scores)
from collections import defaultdict
#if a key is not present defaultdict will not throw error, it will add the key in the dict

rank = defaultdict(int)

for i,sent in enumerate(sent_token):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            rank[i] += freq[w]

print (rank)

#this will not store in sorted order
top_sent = nlargest(4, rank, key=rank.get)
print (top_sent)

#sort and print in ascending order
print ([sent_token[j] for j in sorted(top_sent)])
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.probability import FreqDist
from heapq import nlargest
from collections import defaultdict

def article_summary(text,top_n):
    sent_token = sent_tokenize(text)
    word_token = word_tokenize(text.lower())
    stop_custom=set(stopwords.words('english')+list(punctuation)+["''"])
    word_list=[word for word in word_token if word not in stop_custom]
    freq = FreqDist(word_list)

    rank = defaultdict(int)

    for i,sent in enumerate(sent_token):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                rank[i] += freq[w]
    top_sent = nlargest(top_n, rank, key=rank.get)
    print ([sent_token[j] for j in sorted(top_sent)])



text = ""
with open('Data/NLP/news.txt', 'r') as file:
    text = file.read()
article_summary(text, 4)
```
